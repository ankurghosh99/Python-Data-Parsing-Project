```python
with open ("finaldata.txt", encoding = "utf-8") as f:
    data = f.read()
```

```python
#data
```

```python
chunks = data.split("\n\n")
chunks = [chunk for chunk in chunks if len(chunk)>3]
```

```python
#chunks
```

```python
chunks[0]
```

```python
def parse_chunk(chunk):
    chunk = chunk.strip()
    sep_chunk = chunk.split("\n")
    username = sep_chunk[0]
    no_of_posts = int(sep_chunk[1].split("post")[0].replace(",",""))
    
    no_of_follower = float(sep_chunk[2].split("follower")[0].replace(",","").replace("K","").replace("M",""))
    if "K" in sep_chunk[2]:
        no_of_follower = int(no_of_follower * 1000)
    elif "M" in sep_chunk[2]:
        no_of_follower = int(no_of_follower * 1000000)
    else:
        no_of_follower = (no_of_follower)  
        
    no_of_following = float(sep_chunk[3].split("following")[0].replace(",","").replace("K","").replace("M",""))
    if "K" in sep_chunk[3]:
        no_of_following = int(no_of_following * 1000)
    elif "M" in sep_chunk[3]:
        no_of_following = int(no_of_following * 1000000)
    else:
        no_of_following = int(no_of_following)
        
    name = sep_chunk[4]
    if len(sep_chunk)>5:
        type_of_page = sep_chunk[5]
        bio = "\n".join(sep_chunk[6:])
    else:
        type_of_page = "Unknown"
        bio = ""
    return {"username" : username , "no_of_posts" : no_of_posts , "no_of_follower" : no_of_follower , "no_of_following" : no_of_following , "name" : name , "type_of_page" : type_of_page , "bio" : bio}
print(parse_chunk(chunks[0]))    
```

```python
all_chunks = []
for chunk in chunks:
    parsed_chunk = parse_chunk(chunk)
    all_chunks.append(parsed_chunk)

#print(all_chunks)
```

```python
import json
with open ("finaldata1.json","w") as f:
     json.dump(all_chunks,f,indent = 4)
```

```python
max = 0
for chunk in all_chunks:
    if(max<chunk["no_of_posts"]):
        max = chunk["no_of_posts"]
        chunk_with_max_post = chunk
        people_who_has_max_post = chunk["username"]
        
print(chunk_with_max_post)
print(f"The name of the person is {people_who_has_max_post}")
```

```python
max = 0
for chunk in all_chunks:
    if(max<chunk["no_of_follower"]):
        max = chunk["no_of_follower"]
        chunk_with_max_follower = chunk
        people_who_has_max_follower = chunk["username"]

print(chunk_with_max_follower)
print(f"The name of the person is {people_who_has_max_follower}")
```
